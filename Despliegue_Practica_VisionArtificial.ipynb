```python
#Cargamos librerías principales
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Tratamiento de imágenes
from PIL import Image

# Lectura del directorio de imágenes
import os, sys
```

```python
#Cargamos el modelo histograma de color
from tensorflow import keras
model_deep = keras.models.load_model('model_histogram.h5')
model_deep2 = keras.models.load_model('model_rgb.h5')
#Cargamos el labelencoder

import pickle
filename = 'labelencoder.pkl'

labelencoder = pickle.load(open(filename, 'rb'))
```

```python
# Lectura de la carpeta con las imágenes futuras
from PIL import Image

path = "futuro"
files = os.listdir( path )

mat_hist=[] # Matriz con los histogramas
nombres=[] #Nombres de los archivos

# Se leen las imágenes y se calcula el histograma
for f in files:
  if f.endswith('.jpg'):
    #Nombre de la imagen
    print(f)
    nombres.append(f)
    #Leemos la imagen
    img = Image.open(path+'/'+ f)
    img=img.resize((300,300))
    # Calculamos histograma de color
    hist = img.histogram()
    #Adicionamos el histograma a la matriz
    mat_hist.append(hist)

# Data frame con los histogramas
futuro=pd.DataFrame(mat_hist)
```

```python
#rgb
futuro_rgb = []

for f in files:
  if f.endswith('.jpg'):
    print(f)
    img = Image.open(path+'/'+ f)
    img=img.resize((100,100))
    array_img = np.array(img)
    futuro_rgb.append(array_img)

futuro_rgb
```

```python
#Hacemos la predicción
Y_fut_hist = model_deep.predict(futuro)
Y_fut_hist.round(0)

futuro_rgb = np.array(futuro_rgb)

Y_fut_rgb = model_deep2.predict(futuro_rgb)
Y_fut_rgb.round(0)

Y_fut_hist = np.argmax(Y_fut_hist, axis = 1)
Y_fut_rgb = np.argmax(Y_fut_rgb, axis = 1)
```

```python
print(labelencoder.inverse_transform(Y_fut_hist))
print(labelencoder.inverse_transform(Y_fut_rgb))
```

Comparando los resultados logramos ver que primero el modelo rgb indico que todas las fotos son peras, entendible ya que solo se logro una exactitud del 33%, por el otro lado el histograma logro identificar exitosamente la mitad de las imagenes un poco bajo para la exactitud del 83% que dio en el aprendizaje, pero mucho mejor que el modelo rgb.
