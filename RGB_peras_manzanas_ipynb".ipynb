# CLASIFICACIÓN DE IMÁGENES
1. Preparación de las imágenes
2. División de los datos 70-30
3. Aprendizaje de la Red
4. Evaluación del Modelo
5. Guardar el Modelo

```python
#Cargamos librerías principales
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Tratamiento de imágenes
from PIL import Image

# Lectura del directorio de imágenes
import os, sys
```

```python
# Lectura de la carpeta con las imágenes de perros
path = "manzana"
files = os.listdir( path )

manzana=[]
nombres=[] #Nombres de los archivos



# Se leen las imágenes y se calcula el histograma
for f in files:
  if f.endswith('.jpg'):
    #Nombre de la imagen
    print(f)
    nombres.append('manzana')
    #Leemos la imagen
    img = Image.open(path+'/'+ f)
    img=img.resize((100,100))
    #Adicionamos la imagen
    img_array = np.array(img)
    manzana.append(img_array)


manzana
```

```python
# Lectura de la carpeta con las imágenes de gatos
path = "pera"
files = os.listdir( path )

pera=[]



# Se leen las imágenes y se calcula el histograma
for f in files:
  if f.endswith('.jpg'):
    #Nombre de la imagen
    print(f)
    nombres.append('pera')
    #Leemos la imagen
    img = Image.open(path+'/'+ f)
    img=img.resize((100,100))
    #Adicionamos la imagen
    img_array = np.array(img)
    pera.append(img_array)


pera
```

```python
nombres
```

```python
#Unimos los dos conjuntos de datos
data = np.concatenate((manzana, pera))
```

```python
#Se codifican las categorias de la VARIABLE OBJETIVO

from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
nombres=labelencoder.fit_transform(nombres)

nombres
```

```python
#División 70-30
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(data, nombres, test_size=0.3)
```

```python
from keras.preprocessing.image import ImageDataGenerator
# Crear un generador de imágenes de entrenamiento aumentadas (data augmentation)
datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    horizontal_flip=True
)
```

#3. Aprendizaje con Red Profunda

```python
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Flatten

#Asignar semilla para generar números aleatorios
import tensorflow as tf
tf.random.set_seed(3)

#Arquitectura de la red neuronal profunda
model_deep = Sequential()
model_deep.add(Flatten(input_shape=(100,100,3))) #entradas
model_deep.add(Dense(300, activation='relu')) #capa oculta
model_deep.add(Dense(100, activation='relu')) #capa oculta
model_deep.add(Dense(2, activation='softmax')) #capa de salida (valores de la variable objetivo)

#Aprendizaje
optimizer = keras.optimizers.Adam(learning_rate=0.03)#momentum=0.2
model_deep.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

history=model_deep.fit(datagen.flow(X_train, Y_train,batch_size=1),epochs=50)
```

```python
#Gráfica del error por cada iteración
plt.plot(history.history['loss'])
```

```python
#Evaluación modelo deep

loss, acc = model_deep.evaluate(X_test, Y_test, verbose=0) #30%
print(loss)
print(acc)
```

```python
#Guardamos el modelo
model_deep.save('model_rgb.h5')
```
